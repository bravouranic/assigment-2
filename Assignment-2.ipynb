# User se day number (1–7) lo:

# 7 → Sunday

```python
day_num = int(input("Enter your Day Number (1-7): "))
if day_num == 1:
    print("Monday")
elif day_num == 2:
    print("Tuesday")
elif day_num == 3:
    print("Wednesday")
elif day_num == 4:
    print("Thursday")
elif day_num == 5:
    print("Friday")
elif day_num == 6:
    print("Saturday")
elif day_num == 7:
    print("Sunday")
else:
    print("Wrong Input! Please write in bewteen 1 to 7 number")
```

# User se month number lo:

# 12,1,2 → Winter
# 3,4,5 → Spring
# 6,7,8 → Summer
# warna → Autumn

```python
month = int(input("Month number enter karein (1-12): "))
if month in [12, 1, 2]:
    print("Winter")
elif month in [3, 4, 5]:
    print("Spring")
elif month in [6, 7, 8]:
    print("Summer")
elif month in [9, 10, 11]:
    print("Autumn")
else:
    print("wrong input! Please write in between 1 to 12 numbers.")
```

# Simple calculator:
# User se operator lo (+, -, *, /):

# → addition
# → subtraction
# → multiplication
# / → division
# warna → "Invalid operator"

```python
num1 = float(input("Enter First Number: "))
operator = input("Enter Operator (+, -, *, /): ")
num2 = float(input("Enter Sconde Number: "))
if operator == "+":
    result = num1 + num2
    print(f"Result: {num1} + {num2} = {result}")

elif operator == "-":
    result = num1 - num2
    print(f"Result: {num1} - {num2} = {result}")

elif operator == "*":
    result = num1 * num2
    print(f"Result: {num1} * {num2} = {result}")

elif operator == "/":
    if num2 != 0:
        result = num1 / num2
        print(f"Result: {num1} / {num2} = {result}")
    else:
        print("Error: Kisi bhi number ko zero se divide nahi kiya ja sakta.")

else:
    print("Invalid operator! Please sirf +, -, *, ya / ka istemal karein.")
```
